Представьте, что сервис выбирает регионы для расширения сети пунктов выдачи. Есть данные о городах и справочник регионов. Нужно понять, где много городов среднего и крупного размера, а где население сосредоточено в одном-двух городах.

Что тренируем: `read_csv`, выбор колонок, фильтрацию, `assign`, `cut`, `merge`, `groupby`, `pivot_table`.

Сделайте:

1. прочитайте `towns.csv` и `regions.csv`;
2. приведите названия регионов к общему виду и объедините таблицы;
3. добавьте колонку `population_mln`;
4. разделите города на группы по населению;
5. найдите топ-10 регионов по суммарному городскому населению;
6. постройте сводную таблицу: федеральный округ × группа города;
7. сделайте один график, который помог бы объяснить выбор региона.

```python
towns = pd.read_csv("towns.csv")
regions = pd.read_csv("regions.csv", sep=";", decimal=",").rename(
    columns={
        "Субъект": "region_name",
        "округ": "federal_district",
        "площадь": "region_area",
        "процент": "region_size_percent",
    }
)

region_aliases = {
    "Кабардино-Балкария": "Кабардино-Балкарская Республика",
    "Карачаево-Черкесия": "Карачаево-Черкесская Республика",
    "Кемеровская область": "Кемеровская область - Кузбасс",
    "Тюменская область с ХМАО и ЯНАО": "Тюменская область",
    "Архангельская область включая НАО": "Архангельская область",
    "Республика Северная Осетия — Алания": "Республика Северная Осетия – Алания",
}

regions["region_name"] = regions["region_name"].replace(region_aliases)

towns = towns.assign(population_mln=towns["population"] / 1000)
towns_extended = towns.merge(
    regions,
    how="left",
    on=["region_name", "federal_district"],
    validate="many_to_one",
)

towns_extended.head()
```
